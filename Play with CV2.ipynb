```python
import cv2
cv2.__version__
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#importing image
im= cv2.imread('sample.jpg')
plt.imshow(im)
```

```python
#This is imported as BGR color, Converting it to RGB
im_rgb= cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

```python
#converting to grayscale
im_gray= cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(im_gray, cmap= 'gray')
```

```python
# Plot the three channels of the image
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(im_rgb[:, :, i], cmap = 'gray')
plt.show()
```

```python
# Transform the image into HSV and HLS models
im_hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
im_hls = cv2.cvtColor(im, cv2.COLOR_BGR2HLS)
# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20))
ax1.imshow(im_hsv)
ax2.imshow(im_hls)
plt.show()
```
